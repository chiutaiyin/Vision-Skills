```python
import csv
import json
import os
import nltk
from nltk.stem.snowball import *
from tqdm import *
from collections import Counter, OrderedDict
import string
```

```python
q_counter = Counter()
n_sample = 0
maxlen = 0
```

```python
dataset = open('csv/vqa_skill_typ_train.csv', 'r')
skip_header = dataset.readline()
lines = csv.reader(dataset, delimiter=',')

for one_data in tqdm(lines):
    n_sample += 1
    question = one_data[1]
    question = question.lower()
    tokens = nltk.word_tokenize(question)
    token_len = len(tokens)
    maxlen = max([maxlen,token_len])
    q_counter.update(tokens)
print('number of sample = ' + str(n_sample))
print('max len = ' + str(maxlen))
```

```python
q_word_counts = [x for x in q_counter.items()]
q_word_counts.sort(key=lambda x: x[1], reverse=True)
```

```python
### build vocabulary based on question
vocab = [x[0] for x in q_word_counts if x[1] >= 0]
unk_word = '<UNK>'
vocab = [unk_word] + vocab
vocab = OrderedDict(zip(vocab, range(len(vocab))))
json.dump(vocab, open('utils/word2vocab_vqa.json', 'w'), indent=2)
```
